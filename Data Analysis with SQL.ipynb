```python
from pandasql import sqldf

pysqldf= lambda q:sqldf(q,globals())
```

```python
q='select * from df'
pysqldf(q)
```

```python
df.columns
```

```python
q='select gender,Partner,tenure from df'
pysqldf(q)
```

```python
q='select distinct gender,churn from df'
pysqldf(q)
```

```python
q='select distinct contract from df'
pysqldf(q)
```

```python
q='select count(distinct contract) from df'
pysqldf(q)
```

```python
q='select count(distinct gender) from df'
pysqldf(q)
```

```python
q='select * from df where gender="Female" and Partner="No" limit 5'
pysqldf(q)
```

```python
q='select * from df where gender="Female" and TotalCharges>820'
pysqldf(q)
```

```python
q='select * from df where gender="Female" and InternetService in ("Fiber optic","DSL") limit 5'
pysqldf(q)
```

```python
q='select * from df where gender="Male" and MultipleLines like "N%" limit 5'
pysqldf(q)
```

```python
q='select * from df where not gender="Male" limit 3'
pysqldf(q)
```

## Orderby

```python
q='select * from df order by TotalCharges desc'
pysqldf(q)
```

```python
q='select MonthlyCharges,TotalCharges from df order by TotalCharges,MonthlyCharges desc'
pysqldf(q)
```

```python
q='select PaymentMethod,TotalCharges from df order by TotalCharges,PaymentMethod desc'
pysqldf(q)
```

```python
q='select PaymentMethod,TotalCharges from df order by PaymentMethod,TotalCharges desc'
pysqldf(q)
```

```python
q='select * from df where TotalCharges is not null limit 2'
pysqldf(q)
```
